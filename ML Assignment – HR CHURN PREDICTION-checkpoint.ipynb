```python
#Removing outliers in Pecent.hike.expected.in.CTC data using zscore:

from scipy.stats import zscore


newdf[["Pecent.hike.expected.in.CTC"]].mean()

newdf['balance_Pecent.hike.expected.in.CTC'] = newdf['Pecent.hike.expected.in.CTC']
newdf['balance_Pecent.hike.expected.in.CTC']= zscore(newdf['balance_Pecent.hike.expected.in.CTC'])

condition1 = (newdf['balance_Pecent.hike.expected.in.CTC']>3) | (newdf['balance_Pecent.hike.expected.in.CTC']<-3 )
newdf1 = newdf.drop(newdf[condition1].index, axis = 0, inplace = False)
newdf2 = newdf1.drop('balance_Pecent.hike.expected.in.CTC', axis=1)

#original one 
plt.figure(figsize= (20,15))
sns.boxplot(x= newdf["Pecent.hike.expected.in.CTC"], color='lightblue')

#After outlier treatment using z score

plt.figure(figsize= (20,15))
sns.boxplot(x= newdf2["Pecent.hike.expected.in.CTC"], color='lightblue')
print("We managed to get rid to some extreme outlier shown below. ")
```

```python
#Removing outliers in Pecent.hike.expected.in.CTC data using zscore:

from scipy.stats import zscore


newdf[["Percent.hike.offered.in.CTC"]].mean()

newdf['balance_Percent.hike.offered.in.CTC'] = newdf["Percent.hike.offered.in.CTC"]
newdf['balance_Percent.hike.offered.in.CTC']= zscore(newdf['balance_Pecent.hike.expected.in.CTC'])

condition1 = (newdf['balance_Percent.hike.offered.in.CTC']>3) | (newdf['balance_Percent.hike.offered.in.CTC']<-3 )
newdf1 = newdf.drop(newdf[condition1].index, axis = 0, inplace = False)
newdf2 = newdf1.drop('balance_Percent.hike.offered.in.CTC', axis=1)

#original one 
plt.figure(figsize= (20,15))
sns.boxplot(x= newdf["Percent.hike.offered.in.CTC"], color='lightblue')

#After outlier treatment using z score

plt.figure(figsize= (20,15))
sns.boxplot(x= newdf2["Percent.hike.offered.in.CTC"], color='lightblue')
print("We managed to get rid to some extreme outlier shown below. ")
```

# Model Building

```python
# Import Libraries for model validation
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# import libraries for metrics and reportings
from sklearn.metrics import confusion_matrix
```

```python
x = df2.drop(labels = ["Status"],axis = 1)
x.head()
x.drop(labels = ["Offered.band_E1","Offered.band","Location",'Location_Noida',"LOB","LOB_AXON","Candidate.Source","Candidate.Source_Direct"],axis = 1,inplace = True)
```

```python
y = df2["Status"]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.columns
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
print(model.intercept_)
print(model.coef_)

```

# Prediction
